Target unit: Mt

# RCP

```python
RCP = "RCP"
nmvoc_rcp = pyhector.rcp26.NMVOC_emissions.loc[:2005]
```

# CEDS

```python
CEDS = "CEDS"
# Unit in CEDS: ktNMVOC, Hector: Mt, MAGICC: Mt
ceds_nmvoc = read_ceds_csv(
    root / "data/ceds//NMVOC_CEDS_emissions_by_country_v2016_07_26.csv")
nmvoc_ceds = ceds_nmvoc / 1000  # kt -> Mt
```

# Figure

```python
lw = 3
nmvoc_ceds.plot(label=CEDS, linewidth=lw)
nmvoc_rcp.plot(label=RCP, linewidth=lw)
plt.title("Global NMVOC Emissions [Mt]")
plt.legend()
plt.xlabel("")
plt.savefig(
    root / "content/images/global-nmvoc.png",
    bbox_inches="tight", 
    pad_inches=0
)
```
