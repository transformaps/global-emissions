# RCP

```python
RCP = "RCP"
bc_rcp = pyhector.rcp26.BC_emissions.loc[:2005]
```

# CEDS

```python
CEDS = "CEDS"
# Unit in CEDS: KtBC, Hector: Mt, MAGICC: Mt
ceds_bc = read_ceds_csv(
    root / "data/ceds/BC_CEDS_emissions_by_country_v2016_07_26.csv")
bc_ceds = ceds_bc / 1000  # kt -> Mt
```

# Figure

```python
lw = 3
bc_ceds.plot(label=CEDS, linewidth=lw)
bc_rcp.plot(label=RCP, linewidth=lw)
plt.title("Global BC Emissions [Mt]")
plt.legend()
plt.xlabel("")
plt.savefig(
    root / "content/images/global-bc.png",
    bbox_inches="tight", 
    pad_inches=0
)
```
