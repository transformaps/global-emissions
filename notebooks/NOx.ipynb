```python
import pandas as pd
import pyhector
import matplotlib.pyplot as plt

plt.style.use("ggplot")
plt.rcParams.update({'font.size': 22})
```

```python
from pandas_datapackage_reader import read_datapackage

from util import root, read_ceds_csv
```

Target unit: MtN

# RCP

```python
RCP = "RCP"
nox_rcp = pyhector.rcp26.NOX_emissions.loc[:2005]
```

# CEDS

```python
CEDS = "CEDS"
# Unit in CEDS: ktNO2
ceds_nox = read_ceds_csv(
    root / "data/ceds//NOx_CEDS_emissions_by_country_v2016_07_26.csv")
nox_ceds = ceds_nox / 1000 * 14 / (14 + 32)  # ktNO2 -> MtN, as for N2O
```

# Figure

```python
lw = 3
nox_ceds.plot(label=CEDS, linewidth=lw)
nox_rcp.plot(label=RCP, linewidth=lw)
plt.title("Global NOx Emissions [MtN]")
plt.legend()
plt.xlabel("")
plt.savefig(
    root / "content/images/global-nox.png",
    bbox_inches="tight", 
    pad_inches=0
)
```
